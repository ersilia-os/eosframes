```python
import pandas as pd
import matplotlib.pyplot as plt

# === 1) Load CSV ===
# Replace with your actual filename
df = pd.read_csv("all_ten_scaled_inf.csv")  # assumes the first row contains column names

# Keep the first 10 columns
cols = df.columns[:10]

# Ensure theyre numeric (coerce non-numeric cells to NaN)
data = df[cols].apply(pd.to_numeric, errors='coerce')

# Optional: check you have ~10 rows of data under each column name
print("Rows per column (non-NaN):")
print(data.count())

# === 2) Boxplot ===
plt.figure(figsize=(12, 6))
plt.boxplot([data[c].dropna().values for c in cols], labels=cols, showfliers=True)
plt.title("Scaled inference(10 Random)-Boxplot of First 10 Columns")
plt.xlabel("Columns")
plt.ylabel("Values")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()

# === 3) Violin plot ===
plt.figure(figsize=(12, 6))
vp = plt.violinplot([data[c].dropna().values for c in cols],
                    showmeans=True, showextrema=True, showmedians=False)

# Matplotlib's violinplot doesn't auto-label x-axis; set them:
plt.title("Scaled inference(10 Random)-Violin Plot of First 10 Columns")
plt.xlabel("Columns")
plt.ylabel("Values")
plt.xticks(ticks=range(1, len(cols) + 1), labels=cols, rotation=45, ha="right")
plt.tight_layout()
plt.show()
```
