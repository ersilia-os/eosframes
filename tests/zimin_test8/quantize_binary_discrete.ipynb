```python
!pip install pandas matplotlib
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv("test_fitted_binary.csv", index_col=0)

# === Plot each column as bar chart ===
for col in df.columns:
    plt.figure(figsize=(8,4))
    plt.hist(df[col], bins=range(-127, 128), edgecolor="black")
    plt.title(f"Value counts for {col}")
    plt.xlabel("Value")
    plt.ylabel("Count")
    plt.xlim(-130, 130)
    plt.show()
```
